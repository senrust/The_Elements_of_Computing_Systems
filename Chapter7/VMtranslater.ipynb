```python
import re
import glob
```

```python
class Parser():
    def __init__(self, filename):
        """
        ファイルを開き、空行、コメントを削除してから各行ごとのリストにし、self.linesとして格納
        現在アセンブラが変換するコマンド行をline_ptrとして0に初期化
        """
        with open(filename) as f:
            self.lines = [re.split("//", line)[0].strip() for line in f.readlines()]
            self.lines = [line for line in self.lines if line]
        self.line_ptr = 0
        return None
    
    def hasMoreCommands(self):
        """
        line_ptr行が存在するかの確認
        """
        try:
            self.command = self.lines[self.line_ptr]
            return True
        except:
            return False
        
    def advance(self):
        """
        line_ptr行をコマンドに変換して、asmcommandsに追加、ポインタを進める
        """
        self.line_ptr= self.line_ptr + 1
        return None
    
    def getcommand(self):
        self.commandtype = self.commandType()
        self.arg1 = None
        self.arg2 = None
        if self.commandtype == "C_RETURN":
            pass
        elif self.commandtype == "C_ARITHMETIC":
            self.arg1 = self.command.split()[0]
        else:
            self.arg1 = self.command.split()[1]
            if self.commandtype in ["C_PUSH", "C_POP", "C_FUNCTION", "C_CALL"]:
                self.arg2 = self.command.split()[2]
        return None
    
    def commandType(self):
        """
        その行のコマンドのタイプを判断
        """
        command_splited = self.command.split()
        head_command = command_splited[0]
        if head_command =="push":
            return "C_PUSH"
        elif head_command =="pop":
            return "C_POP"
        elif head_command =="label":
            return "C_LABEL"
        elif head_command == "goto":
            return "C_GOTO"
        elif head_command == "if":
            return "C_IF"
        elif head_command == "function":
            return "C_FUNCTION"
        elif head_command == "return":
            return "C_RETURN"
        else:
            return "C_ARITHMETIC"
    
    
    def arg1(self):
        self.command.split()[1]
        
        if re.search("^@", self.command):
            return "A_COMMAND"
        elif re.search(r"\(.*\)", self.command):
            return "L_COMMAND"
        else:
            return "C_COMMAND"
```

```python
class PushPop():
    def __init__(self):
        return None
    
    def convert(self, commandtype, arg1, arg2):
        self.commandtype = commandtype
        self.arg1 = arg1
        self.arg2 = int(arg2)
        if self.commandtype == "C_PUSH":
            asm_list = self.convert_push()
        else:
            asm_list = self.convert_pop()
        asm = "\n".join(asm_list)
        return asm
    
    def convert_push(self):#要は各領域の値をスタックに入れる動作
        asm_list = []
        if self.arg1 == "constant":#スタックにarg2を入れ、SPに1を追加する
            asm_list.append("@"+str(self.arg2))
            asm_list.append("D=A")
        elif self.arg1 == "this":
            asm_list.append("@THIS")#pointer領域の示すアドレスをデータに入れる
            asm_list.append("D=M")
            asm_list.append("@"+str(self.arg2))#ARG領域にある何番目の変数であるか
            asm_list.append("A=D+A")#そのポインタの値を取り出し、Dに入れる
            asm_list.append("D=M")
        elif self.arg1 == "that":
            asm_list.append("@THAT")
            asm_list.append("D=M")
            asm_list.append("@"+str(self.arg2))
            asm_list.append("A=D+A")
            asm_list.append("D=M")
        elif self.arg1 == "pointer":
            asm_list.append("@"+str(3+self.arg2))#pointer領域の示すアドレスをデータに入れる
            asm_list.append("D=M")
        elif self.arg1 == "temp":
            asm_list.append("@"+str(self.arg2+5))#temp領域のアドレスのデータを入れる
            asm_list.append("D=M")
        elif self.arg1 == "argument":#ARGの指すアドレスを取り出す
            asm_list.append("@ARG")
            asm_list.append("D=M")#Dにおいておく
            asm_list.append("@"+str(self.arg2))#ARG領域にある何番目の変数であるか
            asm_list.append("A=D+A")#ARG領域にarg2番目の値をDにいれ、スタックへ格納
            asm_list.append("D=M")
        elif self.arg1 == "local":
            asm_list.append("@LCL")#LCLの指すアドレスの値を取り出す
            asm_list.append("D=M")
            asm_list.append("@"+str(self.arg2))#LOLの何番目かをarg2が指すので、そのアドレスを計算
            asm_list.append("A=D+A")#取り出した値をAに設定してDに送る。そしてスタックへ格納
            asm_list.append("D=M")
        elif self.arg1 == "static":
            asm_list.append("@"+str(self.arg2+16))#スタティック領域は16番目から始まる
            asm_list.append("D=M")#static領域のarg2番目の値をDに入れる
        asm_list.append("@SP")
        asm_list.append("A=M")
        asm_list.append("M=D")
        asm_list.append("@SP")
        asm_list.append("M=M+1")
        return asm_list
            
        
        
    def convert_pop(self):#要はSPの値を指定の場所に入れるだけ
        asm_list = []
        if segment == "constant":
            asm_list.append("@SP")
            asm_list.append("AM=M-1")
            asm_list.append("D=M") #SPの指すアドレスの値をDに入れる
            asm_list.append("@"+str(self.arg2))#arg2のアドレスにDを入れる(これ使う?)メモリ直接操作か?
            asm_list.append("M=D")
        elif segment == "this":
            asm_list.append("@THIS")#THISポインタの値をDにおいておく
            asm_list.append("D=M")
            asm_list.append("@"+str(self.arg2))
            asm_list.append("D=D+A")   #Thisポインタ領域のarg2番目のアドレスをDに設定
            asm_list.append("@13")#汎用レジスタにDを避難させる
            asm_list.append("M=D")
            asm_list.append("@SP")#SPの指す値を1つ戻してから、AとMに格納
            asm_list.append("AM=M-1")
            asm_list.append("D=M")#SPの指すアドレスの値をDに入れる
            asm_list.append("@13")#汎用レジスタの値をAに設定
            asm_list.append("A=M")
            asm_list.append("M=D")#スタックから取り出した値DをMに入れる
        elif segment == "that":
            asm_list.append("@THAT")
            asm_list.append("D=M")
            asm_list.append("@"+str(self.arg2))
            asm_list.append("D=D+A")   # D now has target address
            asm_list.append("@13")
            asm_list.append("M=D")
            asm_list.append("@SP")
            asm_list.append("AM=M-1")
            asm_list.append("D=M")
            asm_list.append("@13")
            asm_list.append("A=M")
            asm_list.append("M=D")
        elif segment == "pointer":
            asm_list.append("@SP")#SPの値を1つ減らして、そのアドレスが示す値をDに入れる
            asm_list.append("AM=M-1")
            asm_list.append("D=M")
            asm_list.append("@"+str(self.arg2+3))#pointer領域のarg2番目のアドレスにDを入れる
            asm_list.append("M=D")
        elif segment == "temp":
            asm_list.append("@SP")#SPの値を1つ減らして、そのアドレスが示す値をDに入れる
            asm_list.append("AM=M-1")
            asm_list.append("D=M")
            asm_list.append("@"+str(self.arg2+5))#temp領域のarg2番目のアドレスにDを入れる
            asm_list.append("M=D")
        elif segment == "argument":
            asm_list.append("@ARG")
            asm_list.append("D=M")
            asm_list.append("@"+str(self.arg2))
            asm_list.append("D=D+A")   # D now has target address
            asm_list.append("@13")
            asm_list.append("M=D")
            asm_list.append("@SP")
            asm_list.append("AM=M-1")
            asm_list.append("D=M")
            asm_list.append("@13")
            asm_list.append("A=M")
            asm_list.append("M=D")
        elif segment == "local":
            asm_list.append("@LCL")
            asm_list.append("D=M")
            asm_list.append("@"+str(self.arg2))
            asm_list.append("D=D+A")   # D now has target address
            asm_list.append("@13")
            asm_list.append("M=D")
            asm_list.append("@SP")
            asm_list.append("AM=M-1")
            asm_list.append("D=M")
            asm_list.append("@13")
            asm_list.append("A=M")
            asm_list.append("M=D")
        elif segment == "static":#SPの値を1つ減らして、そのアドレスが示す値をDに入れる
            asm_list.append("@SP")
            asm_list.append("AM=M-1")
            asm_list.append("D=M")
            asm_list.append("@"+str(self.arg2+16))#static領域のarg2番目のアドレスにDを入れる
            asm_list.append("M=D")
        #print "  -> ", asm_list
        return asm_list
        
        
class Arithmetic():
    def __init__(self):
        self.tc = 0 #(TRUEX としてジャンプに使用)
        return None
    
    def convert(self, commandtype):
        asm_list = []
        if commandtype =="add":
            asm_list.append("@SP")#SP呼び出し
            asm_list.append("AM=M-1")#SPのアドレスを一つ戻し、AとMに入れる
            asm_list.append("D=M")#取り出した値をDに入れる
            asm_list.append("@SP")#SP呼び出し
            asm_list.append("A=M-1")#SPのアドレスを1つ戻しAに入れる(SPは動かさないでよい)
            asm_list.append("M=D+M")#addされた結果はAのアドレスに保存される
        elif commandtype =="sub":
            asm_list.append("@SP")#SP呼び出し
            asm_list.append("AM=M-1")#SPのアドレスを一つ戻し、AとMに入れる
            asm_list.append("D=M")#取り出した値をDに入れる
            asm_list.append("@SP")#SP呼び出し
            asm_list.append("A=M-1")#SPのアドレスを1つ戻しAに入れる(SPは動かさないでよい)
            asm_list.append("M=M-D")#addされた結果はAのアドレスに保存される
        elif commandtype =="and":
            asm_list.append("@SP")#SP呼び出し
            asm_list.append("AM=M-1")#SPのアドレスを一つ戻し、AとMに入れる
            asm_list.append("D=M")#取り出した値をDに入れる
            asm_list.append("@SP")#SP呼び出し
            asm_list.append("A=M-1")#SPのアドレスを1つ戻しAに入れる(SPは動かさないでよい)
            asm_list.append("M=D&M")#addされた結果はAのアドレスに保存される
        elif commandtype =="or":
            asm_list.append("@SP")#SP呼び出し
            asm_list.append("AM=M-1")#SPのアドレスを一つ戻し、AとMに入れる
            asm_list.append("D=M")#取り出した値をDに入れる
            asm_list.append("@SP")#SP呼び出し
            asm_list.append("A=M-1")#SPのアドレスを1つ戻しAに入れる(SPは動かさないでよい)
            asm_list.append("M=D|M")#addされた結果はAのアドレスに保存される
        elif commandtype =="neg":
            asm_list.append("@SP")#SP呼び出し
            asm_list.append("A=M-1")#SPのアドレスを一つ戻し、Aに入れる
            asm_list.append("M=-M")#そのAの値を反転させて再代入
        elif commandtype =="not":
            asm_list.append("@SP")#SP呼び出し
            asm_list.append("A=M-1")#SPのアドレスを一つ戻し、Aに入れる
            asm_list.append("M=!M")#そのAの値を反転させて再代入
        elif commandtype =="eq":
            asm_list.append("@SP")
            asm_list.append("AM=M-1") #stackpointer down, address down
            asm_list.append("D=M") #D contains y
            asm_list.append("@SP")
            asm_list.append("A=M-1") #address to x
            asm_list.append("D=M-D") #D=x-y
            asm_list.append("M=-1") #set true
            asm_list.append("@TRUE"+str(self.tc))
            asm_list.append("D;JEQ")
            asm_list.append("@SP")
            asm_list.append("A=M-1")
            asm_list.append("M=0")
            asm_list.append("(TRUE"+str(self.tc)+")")
            self.tc=self.tc+1
        elif commandtype == "gt":
            asm_list.append("@SP")
            asm_list.append("AM=M-1") #stackpointer down, address down
            asm_list.append("D=M") #D contains y
            asm_list.append("@SP")
            asm_list.append("A=M-1") #address to x
            asm_list.append("D=M-D") #D=x-y
            asm_list.append("M=-1") #set true
            asm_list.append("@TRUE"+str(self.tc))
            asm_list.append("D;JGT")
            asm_list.append("@SP")
            asm_list.append("A=M-1")
            asm_list.append("M=0")
            asm_list.append("(TRUE"+str(self.tc)+")")
            self.tc=self.tc+1
        elif commandtype == "lt":
            asm_list.append("@SP")
            asm_list.append("AM=M-1") #stackpointer down, address down
            asm_list.append("D=M") #D contains y
            asm_list.append("@SP")
            asm_list.append("A=M-1") #address to x
            asm_list.append("D=M-D") #D=x-y
            asm_list.append("M=-1") #set true
            asm_list.append("@TRUE"+str(self.tc))
            asm_list.append("D;JLT")
            asm_list.append("@SP")
            asm_list.append("A=M-1")
            asm_list.append("M=0")
            asm_list.append("(TRUE"+str(self.tc)+")")
            self.tc=self.tc+1

        asm = "\n".join(asm_list)
        return asm
```

```python
class CodeWriter():
    def __init__(self, asmfile):
        self.asmfile = asmfile
        self.asmcommands = []
        self.PushPop = PushPop()
        self.Arithmetic = Arithmetic()
        
    def setFileName(self, filename):
        self.Parser = Parser(filename)
        return None
        
    def CodeWriter(self):
        self.Parser.getcommand()
        if self.Parser.commandtype in ["C_PUSH", "C_POP"]:
            self.asmcommands.append(self.writePushPop())
        elif self.Parser.commandtype == "C_ARITHMETIC":
            self.asmcommands.append(self.writeArithmetic())
        self.Parser.advance()
            
        
    def writeArithmetic(self):
        return self.Arithmetic.convert(self.Parser.arg1)
        
    def writePushPop(self):
        return self.PushPop.convert(self.Parser.commandtype, self.Parser.arg1, self.Parser.arg2)
    
    def CheckCommand(self):
        return self.Parser.hasMoreCommands()        
        
    def close(self):
        with open(self.asmfile, mode="w") as f:
            f.write('\n'.join(self.asmcommands) + "\n")
        
```

```python
class VMtranslater():
    def __init__(self, dirname):
        self.dirname = dirname
        self.asmname = self.dirname + "/" + self.dirname.split("/")[-1] + ".asm"
        self.vmfilelist = glob.glob(self.dirname + "/*.vm")
        self.CodeWriter = CodeWriter(self.asmname)
        
    def VMtranslate(self):
        for vmfile in self.vmfilelist:
            self.CodeWriter.setFileName(vmfile)
            while self.CodeWriter.CheckCommand():
                self.CodeWriter.CodeWriter()
        self.CodeWriter.close()
```
